```python
from pathlib import Path
import shutil

SRC_DIR = Path("/root/.cache/kagglehub/datasets/datasnaek/mbti-type/versions/1")

DST_DIR = Path("./data")
DST_DIR.mkdir(parents=True, exist_ok=True)

candidates = list(SRC_DIR.glob("*.csv"))
if not candidates:
    raise FileNotFoundError(f"No CSV found in {SRC_DIR}")

src_csv = next((p for p in candidates if p.name.lower() == "mbti_1.csv"), candidates[0])

dst_csv = DST_DIR / src_csv.name

shutil.copy2(src_csv, dst_csv)

print("Saved dataset to:", dst_csv.resolve())
print("Size (MB):", round(dst_csv.stat().st_size / (1024**2), 2))
```
